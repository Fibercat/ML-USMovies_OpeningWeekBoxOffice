去除2019年，因為總票房尚未完全結算
* 電影數量逐年增加，10年增加40%
* 平均總票房金額為1864萬美元，50%的總票房金額少於20萬美元，75%的總票房金額少於532萬美元
* 平均首週票房金額為654萬美元，50%的總票房金額少於4萬美元，75%的總票房金額少於77萬美元
* 首週票房金額平均約佔總票房的40.48%

```python
boxoffice_df = movies_df.loc[movies_df['Release_year'] < 2019, ['id','Release_year','First_week_gross','Total_gross']]
```

```python
boxoffice_df['First_week_gross'] = boxoffice_df['First_week_gross']/10000
boxoffice_df['Total_gross'] = boxoffice_df['Total_gross']/10000
```

```python
movies_qty = boxoffice_df.groupby('Release_year')['id'].count()
movies_qty
```

```python
(movies_qty[2018] - movies_qty[2009]) / movies_qty[2009]
```

```python
(boxoffice_df['First_week_gross'] / boxoffice_df['Total_gross']).mean()
```

* 以每年的總票房來看，2010,2011年相對低，約103億美元，而2012年相對較高，共計136億美元
* 以每年平均票房收入來看，2013年後都相對較低，皆低於1.8億美元
